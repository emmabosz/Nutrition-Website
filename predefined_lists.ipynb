#### Create a predefined list of top 10 ingredients for each nutrient

```python
nutrients = [
    'FAT (g)', 'PROT (g)', 'Carbohydrates (kcal)', 'VITB1 (mg)', 'RIBF (mg)', 
    'VITB6 (mg)', 'VITB12 (µg)', 'VITD (µg)', 'VITE (mg)', 'VITC (mg)', 'VITA (µg)',
    'Calcium (mg)', 'Iron (mg)', 'Magnesium (mg)', 'Phosphorus (mg)',
    'Potassium (mg)', 'Sodium (mg)', 'Zinc (mg)'
]
```

```python
top10_per_nutrient = {}

for nutrient in nutrients:
    top10 = clean_data[['Engelse naam/Food name', nutrient]].sort_values(by=nutrient, ascending=False).head(10)
    top10_per_nutrient[nutrient] = top10

# Print top 10 ingredients for protein as an example
print("Top 10 Protein-rich foods:")
print(top10_per_nutrient['PROT (g)'])
```
