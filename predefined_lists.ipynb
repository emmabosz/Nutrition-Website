#### Create a predefined list of top 10 ingredients for each nutrient

```python
nutrients = [
    'FAT (g)', 'PROT (g)', 'Carbohydrates (kcal)', 'VITB1 (mg)', 'RIBF (mg)', 
    'VITB6 (mg)', 'VITB12 (µg)', 'VITD (µg)', 'VITE (mg)', 'VITC (mg)', 'VITA (µg)',
    'Calcium (mg)', 'Iron (mg)', 'Magnesium (mg)', 'Phosphorus (mg)',
    'Potassium (mg)', 'Sodium (mg)', 'Zinc (mg)'
]
```

```python
top10_per_nutrient = {}

for nutrient in nutrients:
    # Filter out processed / not everyday foods
    filtered_data = clean_data[
        ~clean_data['food_pyramid_group'].isin(['Processed / Not Everyday', 'Irrelevant', 'Other'])
    ]
    
    # Sort descending by the nutrient and take top 10
    top10 = (
        filtered_data[['Engelse naam/Food name', 'food_pyramid_group', nutrient]]
        .sort_values(by=nutrient, ascending=False)
        .head(10)
    )
    
    top10_per_nutrient[nutrient] = top10

# Example: print top 10 protein-rich foods (excluding processed)
print("Top 10 Protein-rich foods:")
print(top10_per_nutrient['PROT (g)'])
```

##### Merging with clean_data to have nutritional value of all ingredients

```python
clean_data = pd.read_csv('clean_data.csv')
recipe_df = pd.read_csv('NEVO2023_v8.0/NEVO2023_v8.0_Recepten_Recipes.csv', sep='|')

recipe_df
```

```python
# Ensure codes are strings without spaces
recipe_df['NEVO-code'] = recipe_df['NEVO-code'].astype(str).str.strip()
recipe_df['Ingr_NEVO-code'] = recipe_df['Ingr_NEVO-code'].astype(str).str.strip()
clean_data['NEVO-code'] = clean_data['NEVO-code'].astype(str).str.strip()

# Merge on NEVO-code first
recipe_nutrients_df = recipe_df.merge(clean_data, on='NEVO-code', how='left', suffixes=('', '_nutrients'))

# Find rows where merge failed (NaNs in nutrient columns)
missing_mask = recipe_nutrients_df['Engelse naam/Food name'].isna()

if missing_mask.any():
    # Merge missing rows on Ingr_NEVO-code
    ingr_merge = recipe_df[missing_mask].merge(
        clean_data,
        left_on='Ingr_NEVO-code',
        right_on='NEVO-code',
        how='left',
        suffixes=('', '_ingr')
    )

    # Fill missing columns in the main DataFrame
    for col in clean_data.columns:
        if col != 'NEVO-code':  # skip merge key
            recipe_nutrients_df.loc[missing_mask, col] = ingr_merge[col].values

# Drop missing rows if any still remain
recipe_nutrients_df = recipe_nutrients_df.dropna(subset=['Engelse naam/Food name'])
```

```python
recipe_nutrients_df
```

##### Calculating total nutritional value per ingredient based on how much is needed for the recipe

```python
recipe_nutrients_df['Ingr_relatieve_hoeveelheid/Ingr_relative_amount'] = (
    recipe_nutrients_df['Ingr_relatieve_hoeveelheid/Ingr_relative_amount']
    .astype(str)  # Ensure it’s string
    .str.replace(',', '.')  # Replace commas with dots
    .astype(float)  # Convert to float
)
```

```python
for col in nutrients:
    recipe_nutrients_df[f'{col}_actual'] = (
        recipe_nutrients_df['Ingr_relatieve_hoeveelheid/Ingr_relative_amount'].astype(float) / 100
    ) * recipe_nutrients_df[col]
    
    recipe_nutrients_df[f'{col}_actual'] = recipe_nutrients_df[f'{col}_actual'].round(3)

recipe_nutrients_df.head()
```

```python
columns = [
    "Food name", "Ingr_Food", "Food group", "food_pyramid_group"
]
nutrient_actual_cols = [col for col in recipe_nutrients_df.columns if "_actual" in col]
columns += nutrient_actual_cols
```

```python
recipe_final_df = recipe_nutrients_df[columns].groupby("Food name").agg(list).reset_index()

recipe_final_df
```
