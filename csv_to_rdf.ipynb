```python
#imports
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#load data
data = pd.read_csv('NEVO2023_8.0.csv', delimiter="|")

#reduce to relevant variables
variables = ['NEVO-code', 'Food group', 'Engelse naam/Food name', 'FAT (g)', 'PROT (g)', 'ENERCC (kcal)',  'VITB12 (µg)', 'VITD (µg)', 'VITC (mg)', 'FE (mg)' ]
clean_data = data[variables]

#add vegetarian label
clean_data['Vegetarian'] = clean_data['Food group'].apply(lambda x: 'no' if x in ['Meat and poultry', 'Fish, crustacean and shellfish', 'Cold meat cuts'] else 'yes')


wheel_mapping = {
    # Drinks
    'Non-alcoholic beverages': 'Drinks', 

    # Vegetables 
    'Vegetables': 'Vegetables',

    # Fruits
    'Fruit': 'Fruit',

    # Bread, Grain/Cereal Products and Potatoes
    'Cereals and cereal products': 'Bread_Grain_Potatoes',
    'Bread': 'Bread_Grain_Potatoes',
    'Potatoes and tubers': 'Bread_Grain_Potatoes',

    # Dairy
    'Milk and milk products': 'Dairy',
    'Cheese': 'Dairy', 

    #Nuts
    'Nuts and seeds': 'Nuts',

    #Fish, Legumes, Meat and Eggs
    'Meat and poultry': 'Fish_Legumes_Meat_Eggs',
    'Fish, crustacean and shellfish': 'Fish_Legumes_Meat_Eggs',
    'Eggs': 'Fish_Legumes_Meat_Eggs',
    'Legumes': 'Fish_Legumes_Meat_Eggs',
    'Meat substitutes and dairy substitutes': 'Fish_Legumes_Meat_Eggs',
    'Cold meat cuts': 'Fish_Legumes_Meat_Eggs',

    # Spreading and Cooking Fats
    'Fats and oils': 'Spreading and cooking fats',
    'Savoury bread spreads': 'Spreading and cooking fats',

    # Outside the Wheel (too processed, high sugar/salt/fat)
    'Pastry and biscuits': 'Other',
    'Savoury snacks': 'Other',
    'Sugar, sweets and sweet sauces': 'Other',
    'Savoury sauces': 'Other',  
    'Mixed dishes': 'Other',
    'Soups': 'Other',
    'Miscellaneous foods': 'Other',
    'Foods for special nutritional use': 'Other',
    'Herbs and spices': 'Other' ,
    'Alcoholic beverages': 'Other', 
}

# Apply mapping
clean_data['food_group'] = clean_data['Food group'].map(wheel_mapping)

#drop food group other and foods that dont have a food group
clean_data = clean_data[clean_data["food_group"] != "Other"]
clean_data = clean_data.dropna(subset=["food_group"])

#drop the old food group column
clean_data = clean_data.drop(columns=["Food group"])


#rename nutri columns 
clean_data.rename(columns={"ENERCC (kcal)": "Calories (kcal)"}, inplace=True)
clean_data.rename(columns={"FE (mg)": "Iron (mg)"}, inplace=True)
clean_data.rename(columns={"PROT (g)": "Protein (mg)"}, inplace=True)
clean_data.rename(columns={"VITB12 (µg)": "Vitamin B12 (µg)"}, inplace=True)
clean_data.rename(columns={"VITD (µg)": "Vitamin D (µg)"}, inplace=True)
clean_data.rename(columns={"VITC (mg)": "Vitamin C (mg)"}, inplace=True)
clean_data.rename(columns={"Engelse naam/Food name": "Food item"}, inplace=True)





clean_data
```

```python
#save as csv
clean_data.to_csv("foods_for_ontology.csv", index=False)
```
